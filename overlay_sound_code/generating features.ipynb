```python
import pandas as pd
import numpy as np
import os
import librosa
```

```python
meta=pd.DataFrame(os.listdir('gf/')).set_axis(['wavfile'],axis=1)
meta['label']=[1]*meta.size
```

```python
meta1=pd.DataFrame(os.listdir('gv/')).set_axis(['wavfile'],axis=1)
meta1['label']=[2]*meta1.size
```

```python
meta2=pd.DataFrame(os.listdir('fv/')).set_axis(['wavfile'],axis=1)
meta2['label']=[3]*meta2.size
```

```python
def features_extractor(file):
    audio, sample_rate = librosa.load(file, res_type='kaiser_fast') 
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)
    
    return mfccs_scaled_features
```

```python
extracted_features=[]
```

```python
for i in range(meta.shape[0]):
    data=features_extractor('gf/'+meta.at[i,'wavfile'])
    extracted_features.append([data])
    data=features_extractor('gv/'+meta1.at[i,'wavfile'])
    extracted_features.append([data]) 
    data=features_extractor('fv/'+meta2.at[i,'wavfile'])
    extracted_features.append([data])    
    
```

```python
extracted_features=np.array(extracted_features)
```

```python
extracted_features=extracted_features.reshape(3*meta.shape[0],40)
```

```python
extracted_features=pd.DataFrame(extracted_features)
```

```python
a=np.array([1,2,3])
```

```python
a=np.tile(a,50)
```

```python
extracted_features['label']=a
```
